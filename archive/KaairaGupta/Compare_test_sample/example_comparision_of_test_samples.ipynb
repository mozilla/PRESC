```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

data = pd.read_csv("../../../datasets/eeg.csv")

x_features= data.drop(["Class"],axis=1)
x_labels=data["Class"]

x_train,x_test,y_train,y_test = train_test_split(x_features,x_labels,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
logreg.fit(x_train, y_train)
pred_logreg = logreg.predict_proba(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

rdf = RandomForestClassifier()
rdf.fit(x_train, y_train)
pred_rdf = rdf.predict_proba(x_test)
```

```python
import compare_model_classification as cmc

# comparing models for first 20 test cases
cmc.compare_model_classification(pred_logreg[:20], pred_rdf[:20], np.array(y_test-1)[:20])
```
