Example showing usage of the function `visualise_evaluation_metric` in `visualise_evaluation_metric.py`:

```python
import numpy as np
```

Format of input data:
[[x1,x2,x3,...,xn],
 [y1,y2,y3,...,yn],
 .
 .
 .
 [y1,y2,y3,...,yn]
]

Data must be numpy array.

```python
data1 = []  # (exponential distribution)
data2 = []  # (normal distribution)

x = []

# n values of x
n = 10

# create array for x
x = np.arange(n)

# for each x, k values of y
k = 20

# adding rows as [x, y1, y2, y3, ...yn] in data1 (exponential distribution)
for i in range(n):
    r = []
    r.append(x[i])

    a = np.random.exponential(scale=1.0, size=k)
    r.extend(a)

    data1.append(r)

# adding rows as [x, y1, y2, y3, ...yn] in data2 (normal distribution)
for i in range(n):
    r = []
    r.append(x[i])

    a = np.random.normal(scale=1.0, size=k)
    r.extend(a)

    data2.append(r)
```

```python
import visualise_evaluation_metric as vem

# Exponential Distribution
vem.visualise_evaluation_metric(data1)
```

```python
# Normal Distribution
vem.visualise_evaluation_metric(data2)
```
