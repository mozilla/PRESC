### Objective
This notebook aims to study and compare the Reciever Operating Characteristic (ROC) curve on various classificaion models.


### AUC-ROC Curves

AUC–ROC curve is the model selection metric for bi–multi class classification problem. ROC is a probability curve for different classes. ROC tells us how good the model is for distinguishing the given classes, in terms of the predicted probability.

ROC curves features `true positive rate` on the `Y axis`, and `false positive rate` on the `X axis`. This means that the top left corner of the plot is the “ideal” point - a false positive rate of zero, and a true positive rate of one.<br>
This is not very realistic, but it does mean that a **larger area under the curve (AUC) is usually better.**<br>

The **steepness** of ROC curves is also important, since it is ideal to **maximize the true positive rate while minimizing the false positive rate.**<br>

ROC curves are typically used in binary classification to study the output of a classifier. <br>
But they can be extended to multi-label classification, by macro and micro averaging. 


```python
from auc_roc import auc_roc
from classifiers import Classifier
from dataloader import train_val_test_split_data
```

```python
# dividing dtaset into training and test
x_train, x_test, _, y_train, y_test, _ = train_val_test_split_data(0.2)
```

### Interpretation of the Graph
The ROC curve of a random classifier with the random performance level (as shown below) always shows a straight lineThis random classifier ROC curve is considered to be the baseline for measuring the performance of a classifier. <br>Two areas separated by this ROC curve indicates an estimation of the performance level—good or poor.
![title](https://intellipaat.com/blog/wp-content/uploads/2019/06/python6.jpg)
ROC curves that fall under the area at the *top-left* corner indicate *good performance* levels, whereas ROC curves fall in the other area at the *bottom-right* corner indicate *poor performance* levels.<br> An ROC curve of a perfect classifier is a combination of two straight lines both moving away from the baseline towards the top-left corner.
![title](https://intellipaat.com/blog/wp-content/uploads/2019/06/python7.jpg)


*Note : For a binary calssifier computing a single ROC curve will be sufficient, the AUC for each ROC will be the same. Multiple ROCs are usually constructed for multi-classification models*

* ##### Gassian Naive Bayes

```python
model = Classifier()
clf = model.Gaussian(x_train, y_train)
y_score = clf.predict(x_test)

auc_roc(y_test, y_score)
```

* ##### KNieghbors Classifier

```python
clf = model.KNeighbors(x_train, y_train)
y_score = clf.predict(x_test)

auc_roc(y_test, y_score)
```

* ##### SVM Classifier

```python
clf = model.svm_classifier(x_train, y_train)
y_score = clf.predict(x_test)

auc_roc(y_test, y_score)
```

* ##### Decision Tree Classifier

```python
clf = model.Decision_Tree(x_train, y_train)
y_score = clf.predict(x_test)

auc_roc(y_test, y_score)
```

* ##### Random Foreset 

```python
clf = model.Random_Forest(x_train, y_train)
y_score = clf.predict(x_test)

auc_roc(y_test, y_score)
```

* ##### Logistic Regression

```python
clf = model.Logistic_Reg(x_train, y_train)
y_score = clf.predict(x_test)

auc_roc(y_test, y_score)
```
