### Graphs
#### 1. Number of Folds Vs Evaluation Metrics
This method is also known as Rotation estimation or out-of-sample testing.The models performances is visualised by recording :
* **Fit Time :** Time the model takes to fit across different number of cross-validation folds.
* **Training Score :** The cross-validation  score for training data. 
* **Test Score :** The cross-validation score for test data. 

The difference between train and test score can be an indication of overfittness or variance issues.

#### 2. Repition of Cross-Validation Vs Evaluation Metrics
This method, also known as Monte Carlo cross-validation,creates multiple random splits of the dataset into training and validation data For each such split, the model is fit to the training data, and the following evaluation metrics measured.
* Accuracy
* Precision
* Recall
