### Gain and Lift charts for various models
Lift and Gain Charts are a useful way of visualizing how good a predictive model. They measure how much better one can expect to do with the predictive model comparing without a model. It can be helpful to find the best predictive model among multiple challenger models.
* **Gain :**
Gain at a given decile level is the ratio of cumulative number of targets (events) up to that decile to the total number of targets (events) in the entire data set.
* **Lift :**
It measures how much better one can expect to do with the predictive model comparing without a model. <br>
It is the ratio of gain % to the random expectation % at a given decile level. 
The random expectation at the xth decile is x%.<br>

***Note :*** *Gain and Lift Chart calculate the same thing, the difference lies that a lift chart calculates the gain for random bins of data.*

#### Interpretation of the Graphs
**Gain :** <br>
The minimum percentage of samples required to attain an ideal accuracy rate. The baseline describes the ideal situation of accuracy growth with more percentage of samples. The closer to the baseline lower is the model's performance and vice-versa<br>
**Lift :** <br>
The farther away the actual-line (blue) from the predicted-line(orange), the less is the probability of our prediction to being correct and weaker the model. 
Similarly closer the actual-line (blue) to the predicted-line (orange), the more is the probanility of our prediction being correct.
