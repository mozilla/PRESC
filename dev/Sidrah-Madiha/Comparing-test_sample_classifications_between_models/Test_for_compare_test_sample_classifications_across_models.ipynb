```python
from sklearn import datasets
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split
import compare_test_sample_classifications as compare_testsample
import pandas as pd
import warnings
%matplotlib inline
```

```python
data = pd.read_csv("../../../datasets/defaults.csv")

x_features = data.drop(["defaulted"], axis=1)
x_labels = data["defaulted"]

X_train, X_test, y_train, y_test = train_test_split(x_features, x_labels, test_size=0.3)
```

```python
lr = LogisticRegression()
gnb = GaussianNB()
svc = LinearSVC(C=1.0)
rfc = RandomForestClassifier(n_estimators=100)
```

```python
warnings.filterwarnings('ignore')
warnings.simplefilter('ignore')
compare_testsample.compare_test_sample_classification_for_models([lr, gnb, svc, rfc ],
                                              [ 'Logistic', 'Naive Bayes', 'Support Vector Classification',  'Random Forest' ], 
                                              X_train, y_train, X_test, y_test)


```

## Interpretation of Graph 1:
The first graph shows the spread of class probablities of each model, with the help of this graph we can see SVC has most items that has prediction near 0.8. We can also see Naive Bayes tends to push most of the probablities towards 0.0-0.2 and 0.8-1.0, and thus this model seems more confident in its prediction.
