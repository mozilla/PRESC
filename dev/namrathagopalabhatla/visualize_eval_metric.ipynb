```python
import warnings
from visualize_eval_metric import generate_data, visualize_eval_metric
```

# Generate data
The function ***generate_data*** is used to generate data with different characteristics as per the provided parameters to conduct different experiments.

```python
gaussian = generate_data(n=100, dist='gaussian', scale=0.5)
rayleigh = generate_data(n=100, dist='rayleigh', scale=0.5)
exponential = generate_data(n=100, dist='exponential', scale=0.5)
logistic = generate_data(n=100, dist='logistic', scale=0.5)
binomial = generate_data(n=100, dist='binomial', prob=0.2)
random = generate_data(n=100, dist='random')
```

# Visualize bands
The function ***visualize_eval_metric*** is used to visualize the data provided in the tablular format. Although the data generated and used here are random, it could be done with the help of actual evaluation metric over multiple iterations of an algorithm or, for example a K-Fold validation results. Metric such as accuracy could be plotted this way to compare the variation in the metric over different scenarios, in the above example, it would be different subsets of data in the K-fold validation.

```python
visualize_eval_metric(gaussian)
```
