#**Visualization for evaluation metric**

Importing important libraries:

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Let's first make a table to test our data, where columns are x,y1,y2...y10 and there are 50 datapoints

```
table = np.random.randn(50,11)
for i in range(50):
  table[i,0] = i
```

Now let's plot average y vs x

```
x = (table[:, 0]).astype(int)
avg = np.zeros(50)
Min = np.zeros(50)
Max = np.zeros(50)
for i in range(50):
  avg[i] = np.mean(table[i][1:])  
  Min[i] = np.min(table[i][1:])
  Max[i] = np.max(table[i][1:])
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, avg, label="Avg", linewidth=5)
ax.plot(x, Min, label="Min", linewidth=1)
ax.plot(x, Max, label="Max", linewidth=1)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()  
```
