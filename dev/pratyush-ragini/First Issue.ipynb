```python
from funcs import get_features, split_data, train_perceptron, train_decTree, train_neuralNet, train_svm, scores
```

```python
X,Y = get_features('vehicles.csv')
```

The above function extracts features and target labels from the data, and represent it as a matrix and array respectively. The target labels have been redefined as [0, 1, 2, 3] for [opel, saab, bus, van]

```python
train_X, train_Y, test_X, test_Y = split_data(X,Y,9)
```

The data is divided into 9 sets, out of which one is used as validation in one iteration. In each following itration, the validation set shifts to the next one with the rest being training sets. This goes on in a round-about fashion.

```python
MODEL_perc = train_perceptron(train_X,train_Y) 
scores(MODEL_perc, test_X, test_Y)
```

Here perceptron is used as a training model. It automatically learns the basis functions to be used and also uses a nonlinear function to best fit the data. It's like a neural network with one hidden layer. No regularisation has been performed.

The scores function tells the true estimation rate across classes and averages it across the 9 iterations done on different test sets created. Also it gives the confusion matrix, that too averaged over the 9 iterations (hence the floating point values in matrix). It can clearly be seen that the highest confusion is between opel and saab, since the 2 * 2 top left submatrix is dense which signifies higher number of false positives and negatives between the two. 

```python
MODEL_decTree = train_decTree(train_X,train_Y)  
scores(MODEL_decTree, test_X, test_Y)
```

Here the decision tree classifier has been used. The splitting on nodes is carried out using the maximum information gain (mutual information and entropy concepts of information theory) . 

Here too, the scores function gives the true positive rate and the confusion matrix as well averaged over the 9 iterations. This gives a better score than perceptron as seen by the true positive rate as well as the 2 * 2 submatrix for opel and saab is now less sparse.

```python
MODEL_neuralNet = train_neuralNet(train_X,train_Y) 
scores(MODEL_neuralNet, test_X, test_Y)
```

This model uses a neural network with 80 hidden layers and 78 nodes per layer. As the non-linearity, it uses the logistic sigmoid function. And to obtain the optimum, it uses the ADAM solver.

The scores function here too gives the true positive rate and the confusion matrix averaged over 9 iterations. The accuracy is better than that of perceptron and almost equivalent to that of decision tree. 

```python
MODEL_svm = train_svm(train_X,train_Y) 
scores(MODEL_svm, test_X, test_Y)
```

This is the SVM classifier, which is sort of like perceptron but uses the margin separation to get better results on test sets.It uses RBF i.e. Radial Basis Function as the Kernel function.

The scores function here too gives the true positive rate and confusion matrix averaged over 9 iterations. This gives a much better performance than on any of the above, as seen clearly in the accuracy and the much sparser 2 * 2 submatrix of opel and saab.
